```python
def extracting_text(url):
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser')
    classes = soup.find_all('div', class_ = 'td-footer-template-wrap')
    for footer_class in classes:
        footer_class.decompose()
    text_data = soup.find_all("p")
    comp_text = []
    for text in text_data:
        comp_text.append(text.text)
    return comp_text, soup.title.text
    
```

```python
def making_textfile(title, text, url_id):
    f = open(str(url_id)+".txt", "w")
    f.write(title)
    f.writelines(text)
        
```

```python

def preprocessed_text(text):
    text = [word for word in text if not word in stopwords.words('english')]
    text = re.sub('[^a-zA-Z0-9]', ' ', str(text))
    text = text.lower()
    text = text.split()
    return text
```

```python
pre_web_text=[]
web_text = []
for i in range(len(input)):
    url = input['URL'][i]
    url_id = input['URL_ID'][i]
    text, title = extracting_text(url)
    text.append(title)
    # making_textfile(title, text, url_id)
    web_text.append(text)
    pre_web_text.append(preprocessed_text(text))
    
```

```python
from nltk.corpus import wordnet
nltk.download('wordnet')

def is_complex_word(word):
    if len(nltk.tokenize.word_tokenize(word)) > 1:
        return 1
    
    synsets = wordnet.synsets(word)
    if len(synsets) > 1:
        return 1
    
    return 0

complex_word_count = []
for i in range(len(input['Preprocessed Text'])):
    count = 0
    for word in input['Preprocessed Text'][i]:
        count += is_complex_word(word)
    complex_word_count.append(count)
input['Complex Count'] = complex_word_count
```

```python
import os
stopword_file = r'/run/media/cool_dude69/Local Disk/Machine learning/Data Files/Blackcoffer assignment/20211030 Test Assignment-20230603T103310Z-001/Black_Coffer_Assignment/StopWords' # Stopword file path
materdictionary = r'/run/media/cool_dude69/Local Disk/Machine learning/Data Files/Blackcoffer assignment/20211030 Test Assignment-20230603T103310Z-001/Black_Coffer_Assignment/MasterDictionary' # Master dictionary file path
stopword = []
for filename in os.listdir(stopword_file):
    f= os.path.join(stopword_file, filename)
    with open(f, "r", encoding="latin-1") as file:
        lines = file.readlines()
        for l in lines[1:]:
            stopword.append(l.split()[-1])

print(len(stopword))

```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from tqdm.notebook import tqdm
def preprocessing(text):
    tokens = nltk.word_tokenize(text)
    transform_text = []
    for token in tokens:
        if token.lower() not in stopword:
            transform_text.append(token.lower())
    return transform_text
```

```python
positive_words_file = r'/run/media/cool_dude69/Local Disk/Machine learning/Data Files/Blackcoffer assignment/20211030 Test Assignment-20230603T103310Z-001/Black_Coffer_Assignment/MasterDictionary/positive-words.txt'
f = open(positive_words_file,"r")
positive_words = f.read()

def positive_score(transform_text):
    
    score = 0
    if transform_text != None:
        for text in transform_text:
            if text in positive_words:
                score += 1
        return score
    return score
```

```python
negative_words_file = r'/run/media/cool_dude69/Local Disk/Machine learning/Data Files/Blackcoffer assignment/20211030 Test Assignment-20230603T103310Z-001/Black_Coffer_Assignment/MasterDictionary/negative-words.txt'
f = open(negative_words_file, "r", encoding="latin-1")
negative_words = f.read()
def negative_score(transform_text):
    score = 0
    if transform_text != None:
        for text in transform_text:
            if text in negative_words:
                score -=1
        return score
    return score
```
