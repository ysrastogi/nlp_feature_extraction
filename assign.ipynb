```python
def extracting_text(url):
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'html.parser')
    classes = soup.find_all('div', class_ = 'td-footer-template-wrap')
    for footer_class in classes:
        footer_class.decompose()
    text_data = soup.find_all("p")
    comp_text = []
    for text in text_data:
        comp_text.append(text.text)
    return comp_text, soup.title.text
    
```

```python
def making_textfile(title, text, url_id):
    f = open(str(url_id)+".txt", "w")
    f.write(title)
    f.writelines(text)
        
```

```python

def preprocessed_text(text):
    text = [word for word in text if not word in stopwords.words('english')]
    text = re.sub('[^a-zA-Z0-9]', ' ', str(text))
    text = text.lower()
    text = text.split()
    return text
```

```python
pre_web_text=[]
web_text = []
for i in range(len(input)):
    url = input['URL'][i]
    url_id = input['URL_ID'][i]
    text, title = extracting_text(url)
    text.append(title)
    # making_textfile(title, text, url_id)
    web_text.append(text)
    pre_web_text.append(preprocessed_text(text))
    
```
